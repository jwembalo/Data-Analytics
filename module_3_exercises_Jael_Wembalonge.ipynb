```python
# Write a Python function called histogram() that takes a string and 
# builds a frequency listing of the characters contained in it. 
# Represent the frequency listing as a Python dictionary and print the in sorted order on 'key'.
# (Source: https://www.hackerrank.com/contests/pycode2015/challenges/dictionary-assignment)

# for a string 'abracadabra' your program should create a dictionary 
# hist={'a': 5, 'b': 2, 'r': 2, 'd': 1, 'c': 1}
# Next, you should sort the dictionary 'hist' by keys and print the contents similar to the following format:

# Letter a  appears  5  time/s
# Letter b  appears  2  time/s
# Letter c  appears  1  time/s
# Letter d  appears  1  time/s
# Letter r  appears  2  time/s
```

```python
# Answer:
mystr = input('Enter a string here: ')
mystr = list(mystr)
counts = dict()
for letter in mystr:
  counts[letter] = counts.get(letter,0) + 1
print(counts)
```

```python
def histogram(mystr):
  mystr = list(mystr)
  hist = dict()
  for letter in mystr:
    hist[letter] = hist.get(letter,0) + 1
  return hist

mystr = input('Enter a string: ')
myhist = histogram(mystr)
for k in sorted(myhist.keys()):
  print('Letter ', k, ' appears ', myhist[k], 'time/s')
```

```python
# A candidate name in the following list votes represents a vote casted to the candidate in an election. 
# Write a Python program to print the name of candidates received the maximum vote. If there is tie, print lexicographically smaller name.

votes = ["john", "johnny", "jackie", "johnny", "john", "jackie", "jamie", "jamie", "john","johnny", "jamie", "johnny", "john"]
```

```python
# Answer:
votes = ["john", "johnny", "jackie", "johnny", "john", "jackie", "jamie", "jamie", "john","johnny", "jamie", "johnny", "john"]
#Create a dictionary to store all candidates and their votes
counts = dict()
for vote in votes:
  counts[vote] = counts.get(vote,0) + 1
#Write a function to retrieve keys by a value, 
#so that we have a list of winners (i.e. candidates who all have the maximum votes)
#to sort later to find the lexicographically smallest name  
def GetKeyByValue(SearchDict, SearchValue):
  KeyList = list()
  ItemList = SearchDict.items()
  for item in ItemList:
    if item[1] == SearchValue:
      KeyList.append(item[0])
  return KeyList
#Find the list of winners (candidates who have the highest number of votes)
WinnerList = GetKeyByValue(counts,max(counts.values()))
print('The candidate who received the maximum votes and has the lexicographically smallest name is', min(WinnerList),'.')
```

```python
# Answer:
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter the website address: ')
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all of the paragraph tags and count the number of them 
count = 0
tags = soup('p')
for tag in tags:
  count = count + 1
print(count)

#Print out texts
import urllib.request, urllib.error, urllib.parse
fhand = urllib.request.urlopen(url)
for line in fhand:
  print(line.decode().strip())


```
